# Basic Terminology

Because data science developed separately in math and computer science and because it is used in so many different fields, there are often multiple terms for the same thing. The definition belows give my preferred term and all the synonyms I can think of.

- def: *domain* - The system/field/problem/business/etc to be studied
- def: *population* - The entire domain (everything we'd like to know)
- def: *sample* - A subset of the population that we have data for
- def: *observation* (aka: *instance*, *observational unit*, *data point*) - One element of the sample, typically stored as a row in the dataset
- def: *variable* - A property of interest, typically represented as a column in the dataset. Many people use "feature" synonymously, but we'll reserve that term for a narrower sense of an input variable into a supervised learner. Variables are classified into the following types:
    - def: *quantitative* (aka: *continuous* or *numerical*) - A variable with a continuous set of values
        - ex: weight, income, length, etc
    - def: *categorical* (aka: *qualitative* or "discrete") - A variable with a discrete set of values, with with these subtypes:
        - def: *ordinal* - A categorical variable where the values have a natural order
            - ex: "class year" has values Freshman / Sophmore / Junior / Senior
        - def: *nominal* - A categorical variable where the values do not have a natural order
            - ex: "party"  has values Republican / Democrat / Libertarian / Green / Independent
- ordinal variables can be treated as quantitative when averages make sense
    - ex: number of children in a household is ordinal, but we understand the meaning of "an average household has 2.3 children"
    - ex: dead/alive is often expressed as ordinal as 0/1, but it is hard to understand the meaning of "an average WWII soldier is 0.7 alive"

- def: *data wrangling* (aka: *cleaning* or *preprocessing*) - Collecting data from sources and preparing it for analysis

- def: *descriptive analytics* (aka: *exploration*) - Analyzing data we have already observed
    - always preceeded by data wrangling

- def: *predictive analytics* - Making predictions about data we have not yet observed (based on data we have observed)
    - always preceeded by descriptive analytics

Most data scientist spend MUCH more time on data wrangling than on analytics. This course lies to you in the sense that we'll spend most our time doing analytics (because that is the topic of the course). Don't be fooled - your career will mostly be spent on wrangling.

# Supervision

- def: [*supervised learner*](https://en.wikipedia.org/wiki/Supervised_learning)
    - distinguishes variables into 2 roles
        - def: *feature* (aka: *input / independent/ predictor *) - A variable that the model uses to make predictions
        - def: *target* (aka: *output / dependent / response*) - A variable the model is trying to predict
    - has access to the true values of targets for observations in the datset
    - attempts to predict the targets using the features in a way that will generalize to not-yet-observed samples
        - def: *regressor* - Target variables are quantitative
            - Ex: predict corn yield based on the rain, temperature, soil Nitrogen contenet, and seed type 
        - def: *classifier* - Target variables are categorical
            - Ex: predict whether a income tax return is "normal" or "fraudulent" based on adjusted gross income, tax liability, and itemized deductions
- def: [*unsupervised learner*](https://en.wikipedia.org/wiki/Unsupervised_learning)
    - does not distinguish features and targets - treats all variables the same (basically)
    - often used before supervised learning for feature engineering
    - looks for hidden patterns (rather than predicting a known target like supervised learners)
    - clustering \& dimensionality reduction

- def: [*Semi-supervised learner*](https://en.wikipedia.org/wiki/Semi-supervised_learning)
    - alternates between supervised and unsupervised
    - common problem: lots of observations, but only a few have been labeled (don't know values of target variables for most samples)
        - ex: income tax returns
            - only a small fraction few get audited
            - sampling bias toward the fraudulent class - audits are not imposed via simple random sample; typically done if something "looks weird" to a human
            - proportion of auditted tax return that are fraudulent is much higher than the proportion of non-auditted tax return that are fraudulent
            - supervised learner only trains on auditted (aka: labeled) tax returns, but gets applied to all tax return
            - flags many normal return as fraudulent (false positives)
        - use iterative training, often with a human in the loop, to guide the model and combat effect of sampling
        
- [*Reinforcement learner*](https://en.wikipedia.org/wiki/Reinforcement_learning)
    - *agent* operating in a environment subject to a reward function or opponent or external force, etc

# Parameters vs Hyperparameters

Models also rely on values NOT in the dataset
- def: *parameter* - Has a well-defined value that fits the training data best and can be algorithmically found by a computer (*training*).
- def: *hyperparameter* - Does not have a well-defined best value.  There is typically a trade-off between competing (un)desirable model characteristics that requires human judgment (*tuning*)
    - There are algorithms to help with this choice (grid search, gradient descent, genetic algorithms, etc), but the responsibility ultimately falls to the data scientist to select hyperparaneter values
- Example
    - one continuous feature x
    - one continuous target y
    - model will fit a curve (regression)
        - $y=\beta_0+\beta_1x$ (line)
        - $y=\beta_0+\beta_1x+\beta_2x^2$ (quadratic)
        - $y=\beta_0+\beta_1x+\beta_2x^2+\beta_3x^3$ (cubic)
        - $y=\beta_0+\beta_1x+\beta_2x^2+\beta_3x^3+\beta_4x^4$ (quartic)
        - $y=\beta_0+\beta_1x+\beta_2x^2+\beta_3x^3+\beta_4x^4+\dots+\beta_kx^k$ (higher)
- $ $
    - polynomial degree $k$ is a hyperparameter - no clear cut best choice
    - polynomial coefficients are parameters - there are values for $\beta_i$'s which best fit the data and we can compute using the [normal equation](https://en.wikipedia.org/wiki/Ordinary_least_squares#Normal_equations)


# Underfitting vs Overfitting

- *underfitting* (aka: *bias*) - model captures too little of the variability in the data and tends to make similar predictions for every oberservatopm
- *overfitting* (aka: *variance*) - model captures too much of the variability in the data and tends to make unstable and inaccurate predictions on new observations
- Using the polynomial regression example above
    - Higher degree ... 
        - Always fits training data better
        - Risks overfitting training data causing poor performance on validations & holdout sets
        - Tuning - Typically, try many different degrees and look at performance/complexity trade-off (ex: AIC, BIC, adjusted $R^2$)
        - Parameters: $\beta's$
            - Training - Once degree is chosen, there are well-defined values for the $\beta's$ that best fit the training data that can be algorithmically found by the computer

```
# from helpers.common_imports import *
# from helpers import utilities as ut

import numpy as np, pandas as pd, matplotlib.pyplot as plt, plotly.express as px
rng = np.random.RandomState(1)
N = 30
sample = pd.DataFrame()
sample['x'] = rng.rand(N) ** 2
sample['y'] = 10 - 1/(sample['x']+0.1) + rng.rand(N) * 0.8
sample.plot('x', 'y', kind='scatter')
```

```
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

sample_predictions = pd.DataFrame()
population_predictions = pd.DataFrame()
sample_predictions['x'] = sample['x'].copy()
population_predictions['x'] = np.linspace(sample['x'].min(), sample['x'].max(), 1000)

reg = Pipeline(steps=(('features', PolynomialFeatures()), ('regressor', LinearRegression())))
S = []
P = []

def predict(X):
    df = X.copy()
    df['y'] = reg.predict(df[['x']])
    df['degree'] = k
    return df

for k in range(20):
    reg.set_params(features__degree=k)
    reg.fit(sample[['x']], sample['y'])
    S.append(predict(sample_predictions))
    P.append(predict(population_predictions))

def viz(L):
    predictions = pd.concat(L)
    fig = px.scatter(predictions, x='x', y='y', animation_frame='degree', color_discrete_sequence=["red"])
    fig.add_traces(px.scatter(sample, x='x', y='y', color_discrete_sequence=["black"]).data)
    fig.update_yaxes(range=[0, 20])
    fig.show()
    return predictions

sample_predictions = viz(S)
```

# Supervised Learning Project Workflow

data_science_flow (1).png

- Split Dataset - We split our observation into subsets for different purposes
    - def: *holdout set*
        - removed immediately
        - kept hidden until the end of model building
        - used to estimate generalization performance (how well model will perform on not-yet-seen samples)
        - not always done if sample size too small (but good practice)
    - def: *modeling set* - observations not in holdout set; further split into ...
        - def: *training set* - observations used to train the model

        - def: *validation / testing set* - observations used to evaluate the model
                
- Model scores
    - def: *training score* - model performance on training set
        - ignore this - its like giving a practice test to a class and then using THE SAME questions as the real test
    - def: *validation score* - model performance on validation set
        - pay attention - it is a reasonable measure of model performance
    - def: *generalization score* - model performance on holdout set
        - pay a lot of attention - it is the most rigorous measure of model's future performance on not-yet-seen samples


- def: *Cross-validation* - split modeling set into training & validation sets, train & test model, repeat.  Gives more robust estimate of model performance & helps improve quality of model selection (tuning) decisions.
 

# Weak/Strong and Lazy/Eager Learners

- Eager vs Lazy Learners - Training a model can take significant computation ...
    - def: An *eager* learner does most of the computations during model building and is therefore quick when applied to not-yet-seen samples (generalization)
    - def: A *lazy* learner must do a significant proportion of its computation when applied to not-yet-seen samples (generalization)

- [*Wisdom of Crowds*](https://en.wikipedia.org/wiki/Wisdom_of_the_crowd) - Often, a bunch of dummies do better than one smartie
    - Sir Francis Galton
    - booth at 1906 Plymouth County Fair
    - "Can you estimate the weight of this ox"
    - Lot's of wild guesses
    - Median within 1% of correct answer
    - Better than experts

- [*Ensemble Methods*](https://en.wikipedia.org/wiki/Ensemble_learning)
    - *weak learners* - Individual machine learners often do not perform well
    - *strong learner* - Combine a bunch of dummies → get a smartie
        - Random forests
        - Bagging
        - Boosting
        - AdaBoosting
        - Gradient Boosting
        - Stacking
